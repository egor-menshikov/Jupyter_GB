```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Задача 1

### 1.1. Скачать данные

```python
df = pd.read_csv('kc_house_data.csv')
```

### 1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(8,6))
sns.histplot(df['price'], bins=80)
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость, млн')
plt.ylabel('Количество');
```

### 1.3. Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(8,6))
sns.histplot(df['sqft_living'], bins=50)
plt.title('Распределение жилплощади')
plt.xlabel('Жилплощадь')
plt.ylabel('Количество');
```

### 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(8,6))
sns.histplot(df['yr_built'], bins=50)
plt.title('Год постройки')
plt.xlabel('Год')
plt.ylabel('Количество');
```

### Выводы:
Большинство домов стоят от 200 до 600 тысяч евро.
Жилая площадь большинства домов составляет от 1000 до 2500 квадратных футов.
Основная масса домов построена начиная с 1950х по наши дни, с бумом строительства с 2000 по 2010.

## 2 задача

### 2.1. Изучите распределение домов от наличия вида на набережную

```python
data_water = df['waterfront'].value_counts()
data_water.index = ['no', 'yes']
```

```python
plt.figure(figsize=(6,4))

plt.pie(data_water,autopct='%1.10f%%')

plt.legend(data_water.index);
```

### Вывод:
Лишь 0.75% от всех домов имеют вид на набережную.

### 2.2. Изучите распределение этажей домов

```python
data_floors = df['floors'].value_counts().reset_index().sort_values(by='floors')
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=data_floors['floors'], y=data_floors['count'])

plt.title('Этажность')
plt.xlabel('Этажи')
plt.ylabel('Количество');
```

### 2.3. Изучите распределение состояния домов

```python
data_condition = df['condition'].value_counts().reset_index().sort_values(by='condition')
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=data_condition['condition'], y=data_condition['count'])

plt.title('Состояние домов')
plt.xlabel('Состояние')
plt.ylabel('Количество');
```
